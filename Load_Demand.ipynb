# Imports

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

# Dataset

```python
dataset = pd.read_csv('Load_demand.csv', error_bad_lines=False)
```

```python
df = dataset.iloc[ : , 1:3]
df = df.drop(288) 
Load = df.iloc[:, 1]
df['Dates'] = pd.to_datetime(df['Date/Time']).dt.date
df['Time'] = pd.to_datetime(df['Date/Time']).dt.time
df = df.drop(['Date/Time'], axis=1)
df = df.drop(['Dates'], axis=1)
print(df)

```

```python
df.plot(x="Time", y=["Total Load"])
plt.show()
```

```python
mean = df['Total Load'].mean()
df_normalzied = df['Total Load']/mean
```

```python
df_normalzied.plot(x="Time", y=["Total Load"])
plt.show()
df_normalzied.to_csv('Load_Demand_Percentage.csv')
```
